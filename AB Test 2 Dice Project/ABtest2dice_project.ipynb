**AB testing 2 Dice**

For this test I will be testing 2 dice to see which one is better. Each dice will have different distribution values for each outcome therefore they are not fair dice and the distribution is not uniform. I will use multiple different approaches to validate the outcome of the tests.

```python
import numpy as np
import random
```

```python
def makedice(inlist1, inlist2):
  import numpy as np

  il2_100  = np.array(inlist2)*100
  problist1 = list(il2_100)
  problist = [int(n) for n in problist1]
  result = []

  for idx in range(len(inlist1)):
    newlist = [inlist1[idx]] * problist[idx]
    result += newlist

  return result

def toss(dicelist):
  import random
  return random.choice(dicelist)


def epGreed_choice(inlist1, inalpha):
  """Return cat 0 90% of the time else choose 10% of the time"""
  import numpy as np
  import random

  c2p = inalpha
  c1p = 1 - inalpha

  inlist2 = [c1p,c2p]

  il2_100  = np.array(inlist2)*100
  problist1 = list(il2_100)
  problist = [int(n) for n in problist1]
  result = []

  for idx in range(len(inlist1)):
    newlist = [inlist1[idx]] * problist[idx]
    result += newlist

  rand_output = random.choice(result)

  if rand_output == 0:
    return 0
  else:
    choice_output = random.choice([0,1])
    return choice_output
```

For Dice 1 the probability for each outcome is as follows:



1.   10%
2.   20%
3.   20%
4.   10%
5.   30%
6.   10%

```python
d1o = [1,2,3,4,5,6]
d1p = [0.1,0.2,0.2,0.1,0.3,0.1]

dice1 = makedice(d1o,d1p)
```

For Dice 2 the probability for each outcome is as follows:



1.   20%
2.   20%
3.   10%
4.   10%
5.   20%
6.   20%

```python
d2o = [1,2,3,4,5,6]
d2p = [0.2,0.2,0.1,0.1,0.2,0.2]

dice2 = makedice(d2o,d2p)
```

Test 1 --- Base Test: Just rolling both dice multiple times

For this test, I will roll each dice multiple times each round. Each round I will sum up the outcomes from each roll from each dice. The dice with the bigger total sum of outcomes will be the better dice.

1. 1st Round = roll 100 times
2. 2nd Round = roll 200 times
3. 3rd Round = roll 500 times
4. 4th Round = roll 1000 times
5. 5th Round = roll 10,000 times
6. 6th Round = roll 100,000 times
7. 7th Round = roll 1,000,000 times

```python
rolls = 100
d1_score = 0
d2_score = 0
for _ in range(rolls):

  d1_score += toss(dice1)
  d2_score += toss(dice2)

print(str(rolls) + ' rolls:')
print('Dice1 total: ', d1_score)
print('Dice2 total: ',d2_score)
```

```python
def rolls(n):

  rolls = n
  d1_score = 0
  d2_score = 0
  for _ in range(rolls):

    d1_score += toss(dice1)
    d2_score += toss(dice2)

  print(str(rolls) + ' rolls:')
  print('Dice1 total: ', d1_score)
  print('Dice2 total: ',d2_score)
```

```python
#Baseline test roll both dice 100,200 ... 1000000 times
roll_list = [100,200,500,1000,10000,100000,1000000]

for rollnum in roll_list:

  print(rolls(rollnum))
  print(' ')
```

Test 1 Conclusion:

After rolling both dice for multiple rounds. It appears that Dice 1 is the better dice. Lets use a epsilon greedy on the next test to see if it validates the outcome from test 1.

Test 2 --- Using a Epsilon Greedy process to find the better dice

For this test I will use an Epsilon Greedy process to find the better dice. Epsilon will be set at 10% which mean if it finds a better dice 90% of the time it will continue to use the better dice (exploit) the other 10% of the the time it will randomly choose between the 2 dice (explore). There will only be 1 round and I will roll 10000 times.

```python
# ****************** Variable Assignment ****************
# epsilon at 10%
random.seed(42)
d1_score = 0
d2_score = 0
number_of_rolls = 10000
currDice = 0
d1roll_index = 1
d2roll_index = 1
epsilon = 0.1

#Randomly choose which dice to roll first
first_roll  = np.random.randint(2)

if first_roll == 0:
  #roll first dice
  currDice = 0
else:
  #roll 2nd dice
  currDice = 1

# ****************** Roll Process ***********************
#roll dice a number of times
for currentRoll in range(number_of_rolls):

  if currDice == 0:
    print('Roll:                   ' + str(currentRoll))
    print('Dice:                   Dice 1')
  else:
    print('Roll:                   ' + str(currentRoll))
    print('Dice:                   Dice 2')

  if currDice == 0:
    current_dice_val = toss(dice1)
    print('Current Dice1 Val:      ' + str(current_dice_val))
    d1_score += current_dice_val
    d1roll_index += 1

  else:
    current_dice_val = toss(dice2)
    print('Current Dice2 Val:      ' + str(current_dice_val))
    d2_score += current_dice_val
    d2roll_index += 1


  print('d1_score:               ' + str(d1_score))
  print('d1n_rolls:              ' + str(d1roll_index))
  d1avg = d1_score/d1roll_index
  print('Current Dice1 avg:      ' + str(d1avg))

  print('d2_score:               ' + str(d2_score))
  print('d2n_rolls:              ' + str(d2roll_index))
  d2avg = d2_score/d2roll_index
  print('Current Dice2 avg:      ' + str(d2avg))

  if d1avg > d2avg:
    currDice = epGreed_choice([0,1],epsilon)
  else:
    currDice = epGreed_choice([1,0],epsilon)

  if currDice == 0:
    print('Next Dice:              Dice 1')
  else:
    print('Next Dice:              Dice 2')

  print(' ')
```

Test 2 Conclusion:

After 10000 rolls, Dice 1 was rolled 9480 times and Dice 2 was rolled 522 times. Since Epsilon Greedy prefers to use the better dice 90% of the time and choose randomly between the 2 dice 10% of the time. Dice 1 being used around 95% of the time mean that it seems Dice 1 is the better dice. Which supports the outcome from test 1. For the next test, I will use an Optimistic Initial Value process to see whether the outcome supports the outcome from Test 1 and 2.  

Test 3 --- Using an Optimistic Initial Value process to find the better dice.

For this test I will set both initial scores at 100. This will cause their averages to be very high at the beginning. Which will force the process to use both dice in an alternating fashion until both averages are closer to their true averages. Then it will prefer to use the better dice 90% of the time and 10% of the time it will randomly choose between the 2. There is only 1 round and I will roll the dice 10000 times.

```python
# ****************** Variable Assignment ****************
# epsilon at 10%
#optimistic initial value, rolls to 100

random.seed(42)

d1_score = 100
d2_score = 100
number_of_rolls = 10000
currDice = 0
d1roll_index = 1
d2roll_index = 1
epsilon = 0.1

#Randomly choose which dice to roll first
first_roll  = np.random.randint(2)

if first_roll == 0:
  #roll first dice
  currDice = 0
else:
  #roll 2nd dice
  currDice = 1

# ****************** Roll Process ***********************
#roll dice a number of times
for currentRoll in range(number_of_rolls):

  if currDice == 0:
    print('Roll:                   ' + str(currentRoll))
    print('Dice:                   Dice 1')
  else:
    print('Roll:                   ' + str(currentRoll))
    print('Dice:                   Dice 2')

  if currDice == 0:
    current_dice_val = toss(dice1)
    print('Current Dice1 Val:      ' + str(current_dice_val))
    d1_score += current_dice_val
    d1roll_index += 1

  else:
    current_dice_val = toss(dice2)
    print('Current Dice2 Val:      ' + str(current_dice_val))
    d2_score += current_dice_val
    d2roll_index += 1


  print('d1_score:               ' + str(d1_score))
  print('d1n_rolls:              ' + str(d1roll_index))
  d1avg = d1_score/d1roll_index
  print('Current Dice1 avg:      ' + str(d1avg))

  print('d2_score:               ' + str(d2_score))
  print('d2n_rolls:              ' + str(d2roll_index))
  d2avg = d2_score/d2roll_index
  print('Current Dice2 avg:      ' + str(d2avg))

  if d1avg > d2avg:
    currDice = epGreed_choice([0,1],epsilon)
  else:
    currDice = epGreed_choice([1,0],epsilon)

  if currDice == 0:
    print('Next Dice:              Dice 1')
  else:
    print('Next Dice:              Dice 2')

  print(' ')
```

Test 3 Conclusion:

After 10000 rolls, Dice 1 was rolled 9125 times and Dice 2 was rolled 877 times. Since it will use the better dice most of the time. The outcome of Test 3 supports the outcomes of test 1 and 2. Supporting the observation that Dice 1 is the better dice.
